```python
import matplotlib.pyplot as plt
import json
plt.rcParams["font.family"] = "cmr10"
plt.rcParams["font.size"] = 20
plt.rcParams["axes.formatter.use_mathtext"] = True
```

```python
def loadData(path):
    with open(path, "r") as f:
        return json.load(f)
    
instance = 16
base_dir = f"./data/inst{instance}"

data = {
    "plain": loadData(f"{base_dir}/vrp-plain-gecode.json"),
    "luby": loadData(f"{base_dir}/vrp-luby-gecode.json"),
    "lns80-gecode": loadData(f"{base_dir}/vrp-lns80-gecode.json"),
    "lns95-gecode": loadData(f"{base_dir}/vrp-lns95-gecode.json"),
    "lns95-sb(6)-gecode": loadData(f"{base_dir}/vrp-lns95-symm_packs-gecode.json"),
}
```

```python
plt.figure(figsize=(7, 4))
plt.plot(data["plain"]["times"], data["plain"]["objs"], "--o", label="plain")
plt.plot(data["luby"]["times"], data["luby"]["objs"], "--o", label="luby")
plt.plot(data["lns80-gecode"]["times"], data["lns80-gecode"]["objs"], "--o", label="lns80")
plt.plot(data["lns95-gecode"]["times"], data["lns95-gecode"]["objs"], "--o", label="lns95")
plt.plot(data["lns95-sb(6)-gecode"]["times"], data["lns95-sb(6)-gecode"]["objs"], "--o", label="lns95-sb")
plt.xscale("log")
plt.xlabel("Time (log-scale)"); plt.ylabel("Objective")
plt.legend()
plt.savefig(f"obj-plot_inst{instance}.pdf", bbox_inches="tight")
```
